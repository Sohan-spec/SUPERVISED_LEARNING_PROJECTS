```python
import numpy as np
import pandas as pd
```

```python
headers_data = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
df=pd.read_csv("dataset_1.data",names=headers_data)
```

```python
df.replace("?",np.nan,inplace=True)
df
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer=SimpleImputer(missing_values=np.nan,strategy="mean")
imputer.fit(df[['normalized-losses','stroke','bore','horsepower','peak-rpm']])
```

```python
missing_data =df.isnull()
```

```python
for i in missing_data.columns.values.tolist():
    print(missing_data[i].value_counts())
    print("")
```

```python
df.dtypes
```

```python
df[['normalized-losses','stroke','bore','horsepower','peak-rpm']]=imputer.transform(df[['normalized-losses','stroke','bore','horsepower','peak-rpm']])
```

```python
imp_mean=SimpleImputer(missing_values=np.nan,strategy="most_frequent")
df[['num-of-doors']]=imp_mean.fit_transform(df[['num-of-doors']])
```

```python
df.dropna(subset=["price"],axis=0,inplace=True)
```

```python
df.reset_index(drop=True,inplace=True)
```

```python
missing_data = df.isnull()
missing_data.head(5)

for column in missing_data.columns.values.tolist():
    print (missing_data[column].value_counts())
    print("") 
```

```python
bins=np.linspace(min(df['horsepower']),max(df['horsepower']),4)
print(bins)
```

```python
group_names=['Low','Medium','High']
```

```python
df['horsepower-bins']=pd.cut(df['horsepower'],bins,labels=group_names,include_lowest=True)
```
