```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe=OneHotEncoder(handle_unknown="ignore",sparse_output=False)
encoded=ohe.fit_transform(df[['ocean_proximity']])
```

```python
encoded_df=pd.DataFrame(encoded,columns=ohe.get_feature_names_out(["ocean_proximity"]))
```

```python
df=pd.concat([df.drop(columns=["ocean_proximity"]),encoded_df],axis=1)
```

```python
df
```

```python
x=df.drop('median_house_value',axis=1)
y=df['median_house_value']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
lasso=Lasso(alpha=0.001)
```

```python
lasso.fit(x_train,y_train)
```

```python
y_pred=lasso.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
print("lasso prediction:",r2_score(y_test,y_pred))
```
