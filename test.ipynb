```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("Book1.csv")
```

```python
cols=["age","salary"]
```

```python
print(df.isna().sum())
```

```python
print(df.isna().mean()*100)
```

```python
df.replace(np.nan,np.nan,inplace=True)
```

```python
missing_vals=df[df.isna().any(axis=1)]
print(missing_vals)
```

```python
from sklearn.impute import SimpleImputer
```

```python
from sklearn.impute import KNNImputer
```

```python
knn_imputer=KNNImputer(n_neighbors=5)
df[cols]=knn_imputer.fit_transform(df[cols])
```

```python
imp_mean=SimpleImputer(missing_values=np.nan,strategy="mean")
df[cols]=imp_mean.fit_transform(df[cols])
```

```python
city=["Delhi","Mumbai","Delhi"]
```

```python
df_encoded=pd.get_dummies(df,columns=["city"])
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder=OneHotEncoder(handle_unknown="ignore",sparse_output=False)
encoded=encoder.fit_transform(df[["city"]])
encoded_df=pd.DataFrame(encoded,columns=encoder.get_feature_names_out(["city"]))
df=pd.concat([df.drop(columns=["city"]),encoded_df],axis=1)
```
